```python
# joining books and users ratings into table
books_with_ratings=ratings.merge(books,on='ISBN')
```

```python
popular_df=books_with_ratings.groupby('Book-Title').agg(num_rating=('Book-Rating','count'),avg_rating=('Book-Rating','mean'))
```

```python
popular_df=popular_df.reset_index()
```

```python
popular_df
```

```python
popular_df.sort_values('num_rating',ascending=False)
```

```python
## popularity is based on the no of people read the book('num_ratings'>300)
#It is based on the rating it get.
popular_df=popular_df[popular_df['num_rating']>300].sort_values('avg_rating',ascending=False)
```

```python
popular_df
```

```python
popular_df=popular_df.head(50)
```

```python
popular_df
```

```python
#For the model deployment I need Book-title, Author, Image URL
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title',
                                                                                  'Book-Author',
                                                                                  'Image-URL-L',
                                                                                  'num_rating', 
                                                                                  'avg_rating']]
```

## Colaborative Filtering

## similar book predication based on user feedback

```python
books_with_ratings.head()
```

```python
# grouping based on user-id tells the no of books rated by each user:
x=books_with_ratings.groupby('User-ID').count()
x
```

```python
x.index
```

```python
x.shape
```

```python
#Select only users who atleast gave feed back for 200 books(power users)
x=x['Book-Rating']>200
x
```

```python
power_users=x[x].index
```

```python
power_users
```

```python
power_users.shape
```

```python
power_users.sort_values()
```

```python
#selecting only records of power users
filtered_ratings=books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
filtered_ratings
```

```python
#iam considering only the best users (atleast  200 books feedback) group them based on the title.
y=filtered_ratings.groupby('Book-Title').count()
y
```

```python
#the above daraframe tell how many users have read the book.
y.sort_values('User-ID',ascending=False)
```

```python
y=y['User-ID']>=50
```

```python
y
```

```python
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
#pivot table giving the ratings for each book from each user
#Book row with user id as column.
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values="Book-Rating")
```

```python
pt
```

```python
pt=pt.fillna(0)
```

```python
pt
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
type(similarity_scores)
```

```python
df_temp=pd.DataFrame(similarity_scores)
```

```python
df_temp
```

```python
pt.index
```

```python
df_temp_name=df_temp
```

```python
df_temp_name.index=pt.index
df_temp_name.columns=pt.index
```

```python
df_temp_name
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])), key= lambda x : x[1], reverse=True)[1:6]
    #lets create emepty list and in that i want at populate with book information 
    #book author book title image url
    #empty list
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```
